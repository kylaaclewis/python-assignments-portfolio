```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent (f, n, start):
    
    def deriv(f, base): 
        #estimate derivative of f(base)
        return(f(base+10**(-5))-f(base))/(10**(-5))


    x_coords = [start] 
    #storing xn's
    y_coords = [f(start)] 
    #storing yn's

    #HERE
    error_bound = 0.000000001
    base = start

    def moveleft(base):
        new = (base - (n*(deriv(f, base))))
        base = new
        x_coords.append(new)
        y_coords.append(f(new))

    def moveright(base):
        new = (base + (n*(deriv(f, base))))
        base = new
        x_coords.append(new)
        y_coords.append(f(new))
         
    for min:
    while abs(deriv(f,base)) > error_bound:
        if deriv(f, base)>0:
            moveleft(base)
        #if the slope is positve (going up),
        #move to the left (down the slope)
        else:
            moveright(base)
       #if the slope is negative (going dpwn),
        #move to the right (down the slope)

    #for max:
    while abs(deriv(f,base)) > error_bound:
        if deriv(f, base)<0:
            moveleft(base)
        #if the slope is positve (going up),
        #move to the left (down the slope)
        else:
            moveright(base)
        #if the slope is negative (going dpwn),
        #move to the right (down the slope)
        
    #if (f(base-n*deriv(base)) <= f(start)
    #
    #####

    plot_range=np.linspace(min(x_coords)-0.5, max(x_coords_+0.5, 10000))
    #set plot range to make look good

    function_range = [f(i) for i in plot_range]
    plt.plot(plot_range, function_range) 
    #plot f(x)
    plt.plot(x_coords, y_coords) 
    #plot sequence xn, f(xn)


    return round(x_coords[-1],3), round(y_coords[-1],3) 
    #return last[-1] xn and yn rounded to (3) decimal places



#set these based on question:

def f(x):
    return x**2
    #return ((-3*x)**3) + ((10*x)**2) - (5*x) -(3)
    #define function!

n = 0.8
#learning rate set

start = 1
#initial point set

gradient_descent(f,n,start)
```
