```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plot

def normal_density(mean, sd, x):
    return (1/(sd * np.sqrt (2*np.pi)))* np.exp(-((x-mean) **2)/(2*sd**2))

def normal_density_graph(means, variances, x_range):
    x_list = np.linspace(x_range[0], x_range[1], 1000)
    for mean, variance in zip(means, variances):
        y_list = []
        for x in x_list:
            y_list.append(normal_density(mean,variance,x))
        plot.plot(x_list, y_list, label=f'Mean: {mean}, Variance: {variance}')
    plot.legend()
    plot.show()

def normal_density_integral(mean, variance, a, b):
    integral, _ = quad(normal_density, a, b, args=(mean, variance))
    return integral

normal_density_graph([0,5,12,14],[1,2,100,0.5],[-10,15,0.01,30])
mean_height = 171
variance_height = 7.12
sd_height = np.sqrt(variance_height)

likelihood = normal_density_integral(mean_height, sd_height, 162, 190)
print("Likelihood of a random man to have a height between 162cm and 190cm:", likelihood)
    
            
        
                    
    
```
